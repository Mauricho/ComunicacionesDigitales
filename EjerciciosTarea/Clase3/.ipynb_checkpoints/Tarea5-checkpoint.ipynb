El modelo de generación de ruido más común en las comunicaciones digitales que resulta en una observación de valores discretos es el canal simétrico binario (del ingles Binary Symmetric Channel o BSC). Este canal se aplica cuando el símbolo de datos de entrada es binario y las probabilidades condicionales se presentan como en la siguiente figura:

La observación ruidosa se diferencia del símbolo de entrada binaria con una probabilidad $p$, que se denomina probabilidad de error o probabilidad de cruce del BSC. 

Suponga que se transmite un símbolo binario  $A \in \{ 0,1 \}$ con probabilidades a priori  $P_A(0)=q$  y  $P_A(1) =1-q$  a través del BSC  descripto arriba. La observación  $Y \in \{ 0, 1 \}$  también es binaria e igual a $A$ con probabilidad $1-p$ .
- Encuentre cualitativamente la regla de decisión ML para el caso de $q= 0.6$ y $p= 0.3$ .
- Encuentre cualitativamente la regla de decisión ML para el caso de $q= 0.5$ y $p= 0.3$ .
- En cual de los casos (punto 1 o punto 2) esta regla de decisión es óptima? dicho en otras palabras, cuando esta regla de decisión produce la minima probabilidad de error?
- Proponga un valor para p y q de forma tal que la regla de decisión ML sea óptima  y se esté en presencia de un canal que muy probablemente invierta el bit transmitido.
- Realice y adjunte la simulación Monte Carlo del problema.

## Resolución

A partir del esquema podemos determinar:

$P_{Y|H}(0|0) = 1-p = 0.7$ Probabilidad de que la salida sea 0 para cuando la entrada es 0

$P_{Y|H}(1|0) = p = 0.3$   Probabilidad de que la salida sea 1 para cuando la hipotesís es 0

$P_{Y|H}(0|1) = p = 0.3$   Probabilidad de que la salida sea 0 para cuando la hipotesís es 1

$P_{Y|H}(1|1) = 1-p = 0.7$ Probabilidad de que la salida sea 1 para cuando la hipotesís es 1



Podemos graficar la distribución de probabilidad para cada hipotesís. Para esto primero importamos las siguientes librerías, que servirán para el siguiente código:

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

A continuación desarrollamos el código para llevar a cabo la simulación de lo antes calculado. Primero la función para cuando $H=0$

```python
def calcular_probabilidad_h0(num_simulaciones):
    
    contadorA = 0       # Probabilidad de que la salida sea 0 para cuando la hipotesís es 0
    contadorB = 0       # Probabilidad de que la salida sea 1 para cuando la hipotesís es 0

    probabilidadA = 0.7 #Probabilidad del evento A
    probabilidadB = 0.3 #Probabilidad del evento B

    for i in range(num_simulaciones):

        #H = 0                    #Hipotesís
        N = random.random()      #Genera un número aleatorio entre 0 y 1
        
        if N < probabilidadA:
            salida = 0           #Evento A
            contadorA += 1
        else:
            salida = 1           #Evento B
            contadorB += 1
            

    probA = contadorA / num_simulaciones    #Probabilidad del evento A
    probB = contadorB / num_simulaciones    #Probabilidad del evento B

    #Datos de la tirada
    etiquetas = ['$Y=0$','$Y=1$']
    datos = [probA,probB]

    #Crear el histograma con etiquetas
    plt.bar(etiquetas,datos,color=['blue','red'])

    return (probA, probB)
```

Gráfico el resultado:

```python
num_simulaciones = 50000 #int(input("Ingrese el número de simulaciones: "))

results = calcular_probabilidad_h0(num_simulaciones)

print(f'La probabilidad de los eventos A: {results[0]} \nLa probabilidad de los eventos B: {results[1]} ')

plt.title('Histograma de la distribución de probabilidad $H=0$')
plt.ylabel('Probabilidad')
plt.show()
```

Ahora para $H=1$

```python
def calcular_probabilidad_h1(num_simulaciones):
    
    contadorA = 0       # Probabilidad de que la salida sea 0 para cuando la hipotesís es 0
    contadorB = 0       # Probabilidad de que la salida sea 1 para cuando la hipotesís es 0

    probabilidadA = 0.3 #Probabilidad del evento A
    probabilidadB = 0.7 #Probabilidad del evento B

    for i in range(num_simulaciones):

        #H = 1                    #Hipotesís
        N = random.random()      #Genera un número aleatorio entre 0 y 1
        
        if N < probabilidadA:
            salida = 0           #Evento A
            contadorA += 1
        else:
            salida = 1           #Evento B
            contadorB += 1
            

    probA = contadorA / num_simulaciones    #Probabilidad del evento A
    probB = contadorB / num_simulaciones    #Probabilidad del evento B

    #Datos de la tirada
    etiquetas = ['$Y=0$','$Y=1$']
    datos = [probA,probB]

    #Crear el histograma con etiquetas
    plt.bar(etiquetas,datos,color=['blue','red'])

    return (probA, probB)
```

Gráfico el resultado:

```python
num_simulaciones = 50000 #int(input("Ingrese el número de simulaciones: "))

results = calcular_probabilidad_h1(num_simulaciones)

print(f'La probabilidad de los eventos A: {results[0]} \nLa probabilidad de los eventos B: {results[1]} ')

plt.title('Histograma de la distribución de probabilidad $H=1$')
plt.ylabel('Probabilidad')
plt.show()
```

## Análisis cualitativo para la determinación del nivel de decisión
Podemos determinar el siguiente criterio de decisión, si es mayor o igual a $0.5$, se esta enviando un 1.
En cambio, si es menor, se esta eviando un cero.

La mínima probabilidad de error se da para el 0.5, ya que es donde tenemos la misma probabilidad de que ocurran ambos eventos.

En el punto 0.5 es donde se produce el mayor error ya que la probabilidad de que se haya transmitido un bit 1 o 0 es la misma. 

## Simulación Monte Carlo

En un primer intento, se realizo la simulación de acuerdo a lo visto en clases anteriores, en donde se aplico un concepto simular utilizando dados. Por lo tanto el código mostrado al principio es adaptado a este ejercicio, en donde se conluye con el gráfico de la cantidad de veces que ocurre cada evento.


Función que me va a permitir calcular la probabilidad a partir del método de Monte Carlo:

```python
def calcular_probabilidad(num_simulaciones):
    
    contadorA = 0       #Contador para el evento A: Y = 0 
    contadorB = 0       #Contador para el evento B: Y = 1

    nivelDecision = 0.5
    
    for i in range(num_simulaciones):

        bit = random.randint(0, 1) #Bit 

        if bit < nivelDecision:        #Si ocurre el evento A
            contadorA += 1
        else:                          #Si ocurre el evento C  
            contadorB += 1

    probA = contadorA / num_simulaciones    #Probabilidad del evento A
    probB = contadorB / num_simulaciones    #Probabilidad del evento B

    #Datos de la tirada
    etiquetas = ['$Y=0$','$Y=1$']
    datos = [probA,probB]

    #Crear el histograma con etiquetas
    plt.bar(etiquetas,datos,color=['blue','red'])

    return (probA, probB)
```

Indico la cantidad de simulaciones, y grafico la cantidad de veces que sucede cada evento.

```python
num_simulaciones = 50000 #int(input("Ingrese el número de simulaciones: "))

results = calcular_probabilidad(num_simulaciones)

print(f'La probabilidad de los eventos A: {results[0]} \nLa probabilidad de los eventos B: {results[1]}')

plt.title('Histograma de 2 bits')
plt.ylabel('Frecuencia')
plt.show()
```

Vemos que los valores obtenidos son aproximados a los calculados de manera teorica, y al sumarlos da 1.
