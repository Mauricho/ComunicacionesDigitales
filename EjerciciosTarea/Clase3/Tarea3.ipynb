## Resolución

### Analisís Matemático

Según lo investigado una regla de decisión es la siguiente: 
\
$$\lambda\geq\theta$$
\
En donde $\theta$ es el umbral, y este toma el siguiente valor:
\
$$\theta = \frac{\lambda_0 + \lambda_1}{2} = \frac{1 + 3}{2} = 2$$
\
Que es el valor medio de los umbrales iniciales. Entonces nuestra regla de decisión sería:     
- Si $\lambda \geq 2$, clasificamos como $\lambda_1= 3$.
- Si $\lambda < 2$, clasificamos como $\lambda_0=1$.

### Simulación

Se importan las siguientes librerías:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

Se definen los valores de cada $\lambda$ y el número de muestras:

```python
cantidad_muestras = 20 #int(input("Ingrese la cantidad de muestras: "))
lambda_real0 = 1
lambda_real1 = 3
```

Función que realizara la distribución de Poisson:

```python
def distribucion_poisson(muestras, lambda_real):
    # Generar valores para x (número de ocurrencias)
    x_values = np.arange(0, muestras + 1)

    # Calcular la probabilidad para cada valor de x
    poisson_probs = poisson.pmf(x_values, mu=lambda_real)

    return x_values, poisson_probs
```

Se calcula la distribución de probabilidad de Poisson:

```python
# Calcular la distribución de probabilidad de Poisson
x_values0, poisson_probs0 = distribucion_poisson(cantidad_muestras, lambda_real0)
x_values1, poisson_probs1 = distribucion_poisson(cantidad_muestras, lambda_real1)
```

Se realiza el gráfico de dicha distribución

```python
# Crear el gráfico de la distribución de probabilidad
plt.bar(x_values0, poisson_probs0, color='skyblue')
plt.bar(x_values1, poisson_probs1, alpha=0.5, color='red')
plt.title('Distribución de Probabilidad de Poisson')
plt.xlabel('Número de Ocurrencias (X)')
plt.ylabel('Probabilidad')
plt.xticks(x_values0)
plt.grid(True)
plt.show()
```

### Conclusión

Podemos notar visualmente que: 
- para $\lambda \geq 2$ corresponde a $\lambda_1 = 3$.
- y para $\lambda < 2$ corresponde a $\lambda_0=1$
