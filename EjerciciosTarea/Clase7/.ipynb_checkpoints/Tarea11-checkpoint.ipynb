## Resolución

Se importan las siguientes librerías:

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

Se generan hipotesís binarias equiprobables. $H = \{0,1\}$.
Y la salida del enconder es $C = \{-1,1\}$

```python
ref = 0.5                                            # Probabilidad de ocurrencia
nSamples = 10000

z = np.random.uniform(size=nSamples)                 # Generador uniforme de números aleatorios
x = [1 if z[i]>ref else -1 for i in range(len(z))]    # Completo el vector binario de acuerdo a la probabilidad de ocurrencia 
```

Se genera la Waveform con unica funcion base $\psi_1(t)=1\{t \epsilon[0,T] \}$. Se supone $T = 16\Delta t$.

```python
T = 16                    # Período de símbolo
for i in x:
    j = 0
    while j < T:
        txSignal = x
        j += 1 
# txSignal = (2*np.repeat(x, 2**4)-1)    #Esta forma es correcta pero no debería utilizarse, no aplico el concepto.
```

Gráficamos la señal para comprobar la salida de la formación de la forma de onda. Primero definimos la función para graficar:

```python
def plotSignal(signal):
    plt.figure(figsize=(20,8), dpi= 80, facecolor='w', edgecolor='k')
    plt.subplot(2,1,1)
    plt.plot(np.real(signal))
    plt.xlabel('n')
    plt.ylabel('Amplitud')
    plt.title('Componente en fase de la señal')
    plt.grid()
```

Ahora graficamos:

```python
plotSignal(txSignal[0:100])
```

Ahora modelamos el ruido Gaussiano, con media cero y varianza ... Definimos la función para modelar el ruido aditivo AWGN.

```python
# Se genera ruido aleatorio con media cero y la varianza especificada 
def add_noise(signal, variance):
    mean = 0
    noise = np.random.normal(mean, np.sqrt(variance), len(signal))
    
    # Se le agrega ruido aditivo a al señal
    noisy_signal = signal + noise
    
    return noisy_signal
```

Obtenemos la señal con ruido, pasando como parámetros la señal y la varianza del ruido: 

```python
variance = 0.1
noisy_signal = add_noise(txSignal, variance)
```

Gráficamos la señal para ver como afecta el ruido:

```python
plotSignal(noisy_signal[0:100])
```

Para obtener la señal realizamos lo siguiente:

```python
T = 16                    # Período de símbolo
for i in x:
    j = 0
    while j < T:
        noisy_signal = x
        j += 1 
```

Gráficamos la señal

```python
plotSignal(noisy_signal[0:100])
```

```python
def psi_1(t, T):
    """
    Función base psi_1(t).
    
    Args:
        t: Tiempo.
        T: Duración de la función base.
        
    Returns:
        Valor de la función base en el tiempo t.
    """
    if 0 <= t <= T:
        return 1
    else:
        return 0
    
# Definir valores de Delta t y T
Delta_t = 0.1  # Por ejemplo, puedes asignar cualquier valor deseado para Delta t
T = 16 * Delta_t

# Crear una secuencia de tiempos t = n * Delta t
num_samples = 100  # Número de muestras
t_values = np.arange(0, num_samples) * Delta_t

# Calcular los valores de la función base para cada tiempo en la secuencia
psi_values = [psi_1(t, T) for t in t_values]

# Imprimir los valores de la función base
print("Valores de la función base psi_1(t):", psi_values)
```
