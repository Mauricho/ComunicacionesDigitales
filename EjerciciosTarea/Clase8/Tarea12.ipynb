Se importan las siguientes librerías:

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

Se generan hipotesís binarias equiprobables. $H = \{0,1\}$.
Y la salida del enconder es $C = \{-1,1\}$

```python
ref = 0.5                                            # Probabilidad de ocurrencia
nSamples = 10000                                     # número de muestras
T = 16                                               # Período de símbolo
rolloff = 0.5                                        # Factor de decaimiento del coseno realzado

z = np.random.uniform(size=nSamples)                 # Generador uniforme de números aleatorios
x = [1 if z[i]>ref else -1 for i in range(len(z))]    # Completo el vector binario de acuerdo a la probabilidad de ocurrencia 
```

El vector $x$ es directamente la salida del encoder.

Se define la función para generar la forma de onda del coseno realzado. Como parámetros se tiene:
    
    - t: Vector de tiempos.
    - T_symbol: Duración del símbolo en muestras.
    - rolloff: Factor de decaimiento del coseno realzado.
    
Y retorna:
    
    - Vector con la forma de onda del coseno realzado.

```python
def raised_cosine(t, T_symbol, rolloff):
    # Calcular el numerador del coseno realzado
    num = np.sin(np.pi * t / T_symbol) * (np.cos(np.pi * rolloff * t / T_symbol))
    
    # Calcular el denominador del coseno realzado
    den = (np.pi * t / T_symbol) * (1 - (2 * rolloff * t / T_symbol)**2)
    
    # Evitar división por cero en el denominador
    den = np.where(t == 0, 1.0, den)
    
    # Calcular la forma de onda del coseno realzado
    waveform = np.where(np.abs(t) <= T_symbol / (2 * rolloff), num / den, 0.0)
    
    return waveform
```

Se genera la forma de onda del coseno realzado.

```python
t = np.arange(0, T, 1)
cosine_waveform = raised_cosine(t - T / 2, T, rolloff)
```

Formación de la señal de transmisión con coseno realzado.

```python
txSignal = np.zeros(nSamples * T)
for i in range(nSamples):
    start_idx = i * T
    txSignal[start_idx:start_idx + T] = x[i] * cosine_waveform
```

Gráficamos la señal para comprobar la salida de la formación de la forma de onda. Primero definimos la función para graficar:

```python
def plotSignal(signal):
    plt.figure(figsize=(20,8), dpi= 80, facecolor='w', edgecolor='k')
    plt.subplot(2,1,1)
    plt.plot(np.real(signal))
    plt.xlabel('n')
    plt.ylabel('Amplitud')
    plt.title('Señal')
    plt.grid()
```

Ahora graficamos:

```python
plotSignal(txSignal[0:800])
```

### Modelado del canal
Ahora escribimos la función para modelar el ruido aditivo AWGN.

Se genera el ruido necesario para implementar una simulación más fiel a la realidad. Se asume una transmisión por aire, la cual tiene características de ruido AWGN.

```python
# Se genera ruido aleatorio con media y varianza especificada 
def add_noise(signal, variance, mean, ideal):
    
    if ideal:          # Si el canal es ideal
        return signal  # No se agrega ruido, se devuelve la señal original

    # Si el canal no es ideal
    noise = np.random.normal(mean, np.sqrt(variance), len(signal))  # Generar ruido gaussiano
    noisy_signal = signal + noise  # Se le agrega ruido aditivo a la señal
    
    return noisy_signal  # Devolver la señal con ruido añadido
```

Obtenemos la señal luego de pasar por el canal. Indicamos como parámetros: la señal a transmitir, la media y la varianza del ruido, y si el canal es o no ideal.

```python
mean = 0          # Media       
variance = 0.1    # Varianza del ruido 
ideal = 0         # Idealización del canal: 1 --> Ideal
                  #                         0 --> No Ideal

noisy_signal = add_noise(txSignal, variance, mean, ideal)
```

Gráficamos la señal para ver como afecta el ruido:

```python
plotSignal(noisy_signal[0:800])
```

### n-Tuple Former del tipo correlador

Se define la función correlador:

```python
# Recibe dos parámetros: Señal e intervalo de tiempo entre los elementos
def correlador(signal,T):
     t = np.arange(0,2+T,T)                 # Vector tiempo

     waveform = np.zeros(len(t))            # Vector forma de onda

     # Generar la forma de onda "cajón" que va entre 1 y 0
     for i in range(len(t)):
        if t[i] <= T:
            waveform[i] = 1
        else:
            waveform[i] = 0
         
     # La correlación se calcula como:
     min_len = len(signal) if len(signal)<len(waveform) else len(waveform)
     corr = 0

     for i in range(min_len):
       corr +=  signal[i] * waveform[i]   

     return corr*T
```

```python
phiPrima = correlador(noisy_signal,T)
print("Correlador: ",phiPrima)
```

```python
rxSignal = []

for i in range(len(noisy_signal)):
    rxSignal.append(noisy_signal[i] * phiPrima)
```

Gráficamos parte de la señal luego de pasar por el n-Tuple Former tipo correlador.

```python
plotSignal(rxSignal[0:800])
```

Como podemos ver, la señal se debe normalizar.

```python
# Convertir la lista a un array de NumPy
rxSignal_array = np.array(rxSignal)

# Calcular la desviación estándar
desviacion = np.std(rxSignal_array)
print("Desviacion:", desviacion)

# Normalizar la señal
rxSignalN = rxSignal_array / desviacion
```

### n-Tuple Former del tipo matched filter

Se define la función Matched Filter:

```python
def matchedFilter(signal, template):
    
    signalInvertida = np.flip(template) # Invertir la señal conocida
    
    y = np.convolve(signal, signalInvertida, mode='valid')
    
    return y
```

```python
base_1 = np.ones(T)       # Base para los 1
base_2 = -1 * np.ones(T)  # Base para los -1  

rxSignal_1 = matchedFilter(noisy_signal, base_1) # Base para los 1
rxSignal_2 = matchedFilter(noisy_signal, base_2) # Base para los -1 
```

```python
rxSignal_combined = rxSignal_2 - rxSignal_1
```

### Decoder criterio MAP para n-Tuple Former del tipo Correlador

Definimos la función para decodificar según el criterio MAP.

```python
def criterioMAP(signal):
    # Calcular la probabilidad de que la señal sea 1 o -1
    prob_1 = np.exp(-0.5 * (np.array(signal) - 1)**2) / np.sqrt(2 * np.pi)
    prob_menos_1 = np.exp(-0.5 * (np.array(signal) + 1)**2) / np.sqrt(2 * np.pi)
    
    # Inicializar la lista para almacenar la señal detectada
    signal_detected = []

    # Comparar las probabilidades de ser 1 y -1 para cada elemento de la señal
    for i in range(len(signal)):
        if prob_1[i] >= prob_menos_1[i]:
            signal_detected.append(1)
        else:
            signal_detected.append(0)

    return signal_detected
```

Realizamos la detección, suponiendo que la palabra código binaria sigue una distribución normal:

```python
señalMAPCorrelador = criterioMAP(rxSignal)
```

```python
plotSignal(señalMAPCorrelador[0:800])
```

### Probabilidad del error obtenido

Se define la función para comparar las señales y poder medir el error:

```python
def Error(señal, fuente, ph_0, ph_1):
    errores_0 = 0
    errores_1 = 0
    
    for i in range(len(fuente)):
        if señal[i] == 0 and fuente[i] == 1:
            errores_1 += 1
        elif señal[i] == 1 and fuente[i] == -1:
            errores_0 += 1
    
    return errores_0/len(fuente) * ph_0 + errores_1/len(fuente) * ph_1
```
