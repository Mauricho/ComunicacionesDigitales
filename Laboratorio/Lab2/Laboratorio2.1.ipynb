Agregamos librerias

```python
#------------------------------- Librerias ------------------------------- 
import numpy as np
import matplotlib.pyplot as plt
```

```python
nSamples     = 2**20                                   #Número de muestras
samplingRate = 2.0e6                                   #Promedio de muestas
frec         = samplingRate/8                          #Frecuencia 
```

A continuación se desarrolla la función para gráficar las señales, ya sea enviadas o recibidas.

```python
def plotSignal(signal):
#------------------------------- Transmitted I component ------------------------------- # Parte Real
# Plot time domain
    plt.figure(figsize=(20,8), dpi= 80, facecolor='w', edgecolor='k')
    plt.subplot(2,1,1)
    plt.plot(np.real(signal))
    plt.xlabel('n')
    plt.ylabel('Amplitud')
    plt.title('Componente en fase de la señal')
    plt.grid()
#------------------------------- Transmitted Q component ------------------------------- # Parte Imaginaria
# Plot time domain
    plt.subplot(2,1,2)
    plt.plot(np.imag(signal))
    plt.xlabel('n')
    plt.ylabel('Amplitud')
    plt.title('Componente en cuadratura de la señal')
    plt.grid()
    plt.show()
```

Y la función para gráficar el espectro de la señal.

```python
def plotSpect(signal,samplingRate):
    plt.subplots(figsize=(15, 3))
    plt.grid()
    plt.title("Power Spectral Density")
    plt.psd(signal, len(signal), samplingRate, color='C1') # Gráfica el espectro de potencia de una señal
    plt.show()
```

## Parámetros de configuración del SDR

Se configura el SDR, como pide la actividad:

```python
#------------------------------- SDR Parameter Configuration -------------------------------
Uri              = Uri
Loopback         = 0             # 0=Disabled, 1=Digital loopback, 2=Retransmiter
SamplingRate     = samplingRate  # Sample rate RX and TX paths[Samples/Sec] #Promedio de muestas                                   

TxLOFreq         = 918e6         # Carrier frequency of TX path [Hz]
TxAtten          = -70           # Attenuation applied to TX path, valid range is -89 to 0 dB [dB]
TxRfBw           = 2.0e6         # Bandwidth of front-end analog filter of TX path [Hz]
RxBufferSize     = nSamples      # Depend on the length of the transmit signal. Maximum tx buffer size 2**24[samples] #Número de muestras
tx_cyclic_buffer = False

RxLOFreq         = TxLOFreq      # Carrier frequency of RX path [Hz]
GainControlModes = "manual"      # Receive path AGC Options: slow_attack, fast_attack, manual
RxHardwareGain   = 70            # Gain applied to RX path. Only applicable when gain_control_mode is set to 'manual'    
RxRfBw           = TxRfBw        # Bandwidth of front-end analog filter of RX path [Hz] 
RxBufferSize     = 2**20         # Rx Buffer size. Maximun value 2**28 [Samples]
```

### Lazo de retorno


```python
sdr.loopback = Loopback
```

### Ancho de banda en RF

```python
sdr.tx_rf_bandwidth = int(TxRfBw)  # Bandwidth of front-end analog filter of TX path [Hz]
sdr.rx_rf_bandwidth = int(RxRfBw)  # Bandwidth of front-end analog filter of RX path [Hz] 
```

### Ganancia del Hardware

```python
sdr.tx_hardwaregain_chan0   = TxAtten # Attenuation applied to TX path, valid range is -90 to 0 dB [dB]
sdr.rx_hardwaregain_chan0   = RxHardwareGain   # Gain applied to RX path. Only applicable when gain_control_mode is set to 'manual'
sdr.gain_control_mode_chan0 = GainControlModes # Receive path AGC Options: slow_attack, fast_attack, manual
```

### Frecuencia de portadora

```python
sdr.tx_lo = int(TxLOFreq)# Carrier frequency of TX path [Hz]
sdr.rx_lo = int(RxLOFreq) # Carrier frequency of RX path [Hz]
```

### Buffers

```python
sdr.rx_buffer_size   = RxBufferSize
sdr.tx_cyclic_buffer = tx_cyclic_buffer
#sdr._ctrl.context.set_timeout(0) # <- disable timeout if you want to use large buffer size
```

### Media y varianza

```python
real = np.real(rxSignal)  #Parte real de la señal
imag = np.imag(rxSignal)  #Parte imaginaria de la señal
```

Calculo la media y la varianza:

```python
# Media:
mean_real = np.mean(real)
mean_imag = np.mean(imag)

#Varianza:
var_real = np.var(real)
var_imag = np.var(imag)

print("Media real:", mean_real)
print("Media imaginaria:", mean_imag)
print("Varianza real:", var_real)
print("Varianza imaginaria:", var_imag)
```

La varianza es una medidad de dispersión que indica que tan dispersos están los valores de un conjunto de datos con respecto a su media. Cuanto mayor sea la varianza, más dispersos estarán los datos, indica una gran variabilidad en la medición.

### Normalización del histograma

Para normalizar un histograma podemos utilizar el parámetro 'density' de la función 'plt.hist()'. Entonces la suma de todas las areas de las barras será igual a 1.
Cada histograma quedara de la siguiente manera:

```python
plt.hist(real,bins=100, density=True) #(datos,cantidad de celdas probabilisticas, normalización)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de la Fase de la Señal')
plt.grid(True)
plt.show()
```

```python
plt.hist(imag,bins=100, density=True) #(datos,cantidad de celdas probabilisticas, normalización)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de la Cuadratura de la Señal')
plt.grid(True)
plt.show()
```

### Histograma normalizado y función de densidad de probabilidad

Para realizar la distribución normal se importara la librería 'scipy'

```python
from scipy.stats import norm, probplot
```

Se calcula la desviacón estandar, se define la distribución normal y se define la función densidad de probabilidad. 

```python
x = np.arange(-40, 40, step=0.5)  #Genera un array dentro de un rango especifico sin incluir los extremos

#Calcula la desviación estandar
std_real = np.std(real)           
std_imag = np.std(imag)

#Defino la distribución normal
normal_real = norm(mean_real, std_real) #(media, desviación estandar)
normal_imag = norm(mean_imag, std_imag)

#Defino la función de densidad de probabilidad
pdf_real = normal_real.pdf(x)
pdf_imag = normal_imag.pdf(x)
```

Histograma de la parte real:

```python
# Histograma parte real
plt.figure(figsize=(10,6))
plt.xlabel("Valor")
plt.ylabel("Histograma de la fase y PDF de una distribución gaussiana")

#Grafico el histograma
plt.hist(real, bins=100, density=True)  #(datos,cantidad de celdas probabilisticas, normalización)

#Gráfico de la función densidad de probabilidad
plt.plot(x, pdf_real,color='red')

plt.grid(True)
plt.show()
```

Histograma de la parte imaginaria:

```python
# Histograma parte imaginaria
plt.figure(figsize=(10,6))
plt.xlabel("Valor")
plt.ylabel("Histograma de la cuadratura y PDF de una distribución gaussiana")

#Grafico el histograma
plt.hist(imag, bins=100, density=True) #(datos,cantidad de celdas probabilisticas, normalización)

#Gráfico de la función densidad de probabilidad
plt.plot(x, pdf_imag,color='red')

plt.grid(True)
plt.show()
```

### Gráfico Q-Q

Creamos el gráfico QQ para la parte imaginaria

Si los puntos estan cerca de la línea significa que sigue una distribución normal.

### Código con misma media y varianza

Mediante un código se genera una variable gaussiana compleja que tenga la misma media y varianza que la señal muestreada. 
Tener en cuenta que la desviación estándar es la raíz cuadrada de la varianza.

```python
gaussian_real = np.random.normal(mean_real, np.sqrt(var_real), len(real))   #(media,desviación estandar, cantidad de datos)
gaussian_imag = np.random.normal(mean_imag, np.sqrt(var_imag), len(imag))
```

Combino las partes real e imaginaria para formar una variable gaussiana compleja

```python
gaussian_complex = gaussian_real + 1j * gaussian_imag
```

Calculo el histograma de la parte real e imaginaria de la variable gaussiana compleja

```python
plt.figure(figsize=(10, 6))

#       (datos,cant. de celdas probabilisticas, normalización, transparencia, etiqueta)
plt.hist(np.real(gaussian_complex), bins=50, density=True, label='Parte Real')
plt.hist(np.imag(gaussian_complex), bins=50, density=True, alpha=0.5, label='Parte Imaginaria')

plt.title('Histograma de Variable Gaussiana Compleja')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')

plt.legend()
plt.grid(True)
plt.show()
```
